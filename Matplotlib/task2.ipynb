С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.
Создайте массивы y1, y2, y3, y4 по следующим формулам:

y1 = x**2

y2 = 2 * x + 0.5

y3 = -3 * x - 1.5

y4 = sin(x)

Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов. В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.Дайте название графикам: 'График y1', 'График y2' и т.д.
Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.


```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)

title_dict = {'fontsize': 10, 'family': 'Times New Roman'}
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title('График y1', fontdict=title_dict )
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График y2', fontdict=title_dict )

ax3.plot(x, y3)
ax3.set_title('График y3', fontdict=title_dict )

ax4.plot(x, y4)
ax4.set_title('График y4', fontdict=title_dict )

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
```

Число бинов должно равняться 20 для обеих гистограмм, а коэффициент alpha сделайте равным 0.5, чтобы гистограммы были полупрозрачными и не загораживали друг друга. Создайте легенду с двумя значениями: “Class 0” и “Class 1”. Гистограмма обычных транзакций должна быть серого цвета, а мошеннических - красного. Горизонтальной оси дайте название “V1”.


```python
import pandas as pd

plt.style.use('fivethirtyeight')
data = pd.read_csv('creditcard.csv')
data.sample(7)
```

```python
data.value_counts(subset='Class').plot(kind="barh", logx=True);
```

```python
v1_thiefs = data[data.Class == 1].V1
v1_custom = data[data.Class == 0].V1

fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()

ax1.hist(v1_thiefs, bins=20, density=True)
ax1.set_title('thiefs')

ax2.hist(v1_custom, bins=20, density=True)
ax2.set_xlim(-20, 20)
ax2.set_title('customers')

fig.set_size_inches(10, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
plt.hist(v1_thiefs, bins=20, alpha=0.5, density=True, label='thiefs')
plt.hist(v1_custom, bins=20, alpha=0.5, density=True, label='ordinary customer')
plt.xlim(-30, 5)
plt.legend(frameon=False);
```
